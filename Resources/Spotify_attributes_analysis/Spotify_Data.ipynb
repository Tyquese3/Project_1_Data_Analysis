```python
#import dependencies
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#create a reference to csv file
file_path = ("../Resources/Spotify_data.csv")
#Read csv file 
spotify_df = pd.read_csv(file_path)
#print first 5 rows
spotify_df.head()
```

```python
#check the names of all columns and check rows for missing data
spotify_df.count()
```

```python
#remove rows with missing data
cleaned_rows_df = spotify_df.dropna(how="any")
cleaned_rows_df.count()
```

```python
# #remove unwanted columns
# reduced_df = cleaned_rows_df[['song_title', 'artist', 'target', 'danceability', 'duration_ms', 'tempo']]
# reduced_df.head()
```

```python
#Identify duplicates
duplicates_df = cleaned_rows_df[cleaned_rows_df.duplicated()]
print("Duplicate Rows:\n", duplicates_df)
```

```python
# Drop duplicates
cleaned_spotify_complete = cleaned_rows_df.drop_duplicates()
cleaned_spotify_complete.head()
```

```python
cleaned_spotify_complete.count()
```

```python
cleaned_spotify_complete.describe()
```

```python
numeric_cols = ["acousticness", "danceability", "energy", "loudness", "speechiness", "liveness", "valence", "tempo", "duration_ms","instrumentalness"]
cleaned_spotify_complete[numeric_cols].hist(bins=20, figsize=(15, 10))
plt.show()
```

```python
# Sample DataFrame (replace with your actual DataFrame)
data = {
    'song': [f'Song{i}' for i in range(2011)],
    'acousticness': [round(i % 10 * 0.1, 1) for i in range(2011)],
    'danceability': [round(i % 10 * 0.1, 1) for i in range(2011)],
    'instrumentalness': [round(i % 10 * 0.1, 1) for i in range(2011)],
    'energy': [round((i % 10) * 0.1, 1) for i in range(2011)],
    'loudness': [-60 + (i % 10) for i in range(2011)],
    'speechiness': [round((i % 10) * 0.1, 1) for i in range(2011)],
    'liveness': [round((i % 10) * 0.1, 1) for i in range(2011)],
    'valence': [round((i % 10) * 0.1, 1) for i in range(2011)],
    'tempo': [120 + (i % 10) for i in range(2011)],
    'duration_ms': [200000 + (i % 1000) for i in range(2011)],
    'target': [1 if i % 2 == 0 else 0 for i in range(2011)],  # Example target
    'artist': [f'Artist{i % 50}' for i in range(2011)]
}
cleaned_spotify_complete = pd.DataFrame(data)

# # Plot histograms for multiple columns
# columns_to_plot = ['acousticness', 'danceability', 'energy', 'loudness', 'speechiness', 'liveness', 'valence', 'tempo', 'duration_ms']
# cleaned_spotify_complete[columns_to_plot].hist(bins=30, figsize=(15, 10), color='skyblue', edgecolor='black')
# plt.suptitle('Histograms of Audio Features')
# plt.show()
```

```python

colors = {0: 'red', 1: 'orange'} 
styles = {0: 'o', 1: 's'} # 'o' for circles, 's' for squares 
plt.figure(figsize=(15, 20))
plt.legend()
# Subplot 1
plt.subplot(3, 2, 1) 
for target, group in cleaned_spotify_complete.groupby('target'):
    plt.scatter(group['acousticness'], group['danceability'],
                color=colors[target], label=f'Target {target}', marker=styles[target])
    plt.xlabel('Acousticness') 
    plt.ylabel('Danceability') 
    plt.title('Scatterplot for acousticness vs danceability')
    plt.legend()
  
# Subplot 2
plt.subplot(3, 2, 2) 
for target, group in cleaned_spotify_complete.groupby('target'):
    plt.scatter(group['duration_ms'], group['energy'],
                color=colors[target], label=f'Target {target}', marker=styles[target])
    plt.xlabel('Duration_ms') 
    plt.ylabel('Energy') 
    plt.title('Scatterplot for duration_ms vs energy')
    plt.legend()
   
# Subplot 3
plt.subplot(3, 2, 3) 
for target, group in cleaned_spotify_complete.groupby('target'):
    plt.scatter(group['instrumentalness'], group['liveness'],
                color=colors[target], label=f'Target {target}', marker=styles[target])
    plt.xlabel('Instrumentalness') 
    plt.ylabel('Liveness') 
    plt.title('Scatterplot for instrumentalness vs  liveness')
    plt.legend()

# Subplot 4
plt.subplot(3, 2, 4) 
for target, group in cleaned_spotify_complete.groupby('target'):
    plt.scatter(group['loudness'], group['speechiness'],
                color=colors[target], label=f'Target {target}', marker=styles[target])
    plt.xlabel('Loudness') 
    plt.ylabel('Speechiness') 
    plt.title('Scatterplot for loudness vs speechiness')
    plt.legend()
# Subplot 5
plt.subplot(3, 2, 5) 
for target, group in cleaned_spotify_complete.groupby('target'):
    plt.scatter(group['tempo'], group['valence'],
                color=colors[target], label=f'Target {target}', marker=styles[target])
    plt.xlabel('Tempo') 
    plt.ylabel('Valence') 
    plt.title('Scatterplot for tempo vs valence')
    plt.legend()

plt.show()
```
