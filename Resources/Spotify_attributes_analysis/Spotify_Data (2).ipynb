```python
#import dependencies
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#create a reference to csv file
file_path = ("../Resources/Spotify_data.csv")
#Read csv file 
spotify_df = pd.read_csv(file_path)
#print first 5 rows
spotify_df.head()
```

```python
#check the names of all columns and check rows for missing data
spotify_df.count()
```

```python
#remove rows with missing data
cleaned_rows_df = spotify_df.dropna(how="any")
cleaned_rows_df.count()
```

```python
# #remove unwanted columns
# reduced_df = cleaned_rows_df[['song_title', 'artist', 'target', 'danceability', 'duration_ms', 'energy']]
# reduced_df.head()
```

```python
#Identify duplicates
duplicates_df = cleaned_rows_df[cleaned_rows_df.duplicated()]
print("Duplicate Rows:\n", duplicates_df)
```

```python
# Drop duplicates
cleaned_spotify_complete = cleaned_rows_df.drop_duplicates()
cleaned_spotify_complete.head()
```

```python
cleaned_spotify_complete.count()
```

```python
#rename columns
renamed_spotify_complete = cleaned_spotify_complete.rename(columns={"song_title": "song", "artist": "singer"})
renamed_spotify_complete.head()
```

```python

# Create a dictionary from the DataFrame columns
data = {
    'song': renamed_spotify_complete['song'].tolist(),
    'acousticness': renamed_spotify_complete['acousticness'].tolist(),
    'danceability': renamed_spotify_complete['danceability'].tolist(),
    'energy': renamed_spotify_complete['energy'].tolist(),
    'loudness': renamed_spotify_complete['loudness'].tolist(),
    'speechiness':renamed_spotify_complete['speechiness'].tolist(),
    'liveness': renamed_spotify_complete['liveness'].tolist(),
    'valence': renamed_spotify_complete['valence'].tolist(),
    'tempo':renamed_spotify_complete['tempo'].tolist(),
    'duration_ms': renamed_spotify_complete['duration_ms'].tolist(),
    'instrumentalness': renamed_spotify_complete['instrumentalness'].tolist(),
    'target': renamed_spotify_complete['target'].tolist(),
    'singer': renamed_spotify_complete['singer'].tolist()
}

# Create DataFrame from the dictionary
spotify_complete = pd.DataFrame(data)
spotify_complete
```

```python
# Filter liked and unliked songs 
liked_songs = spotify_complete[spotify_complete['target'] == 1] 
unliked_songs = spotify_complete[spotify_complete['target'] == 0]
```

```python
singer_counts = liked_songs['singer'].value_counts().head(20)
singer_counts
```

```python
#create bar chart showing popularity of artists
plt.figure(figsize=(8, 8))
singer_counts.plot(kind='bar', color='lightgreen') 
plt.xlabel('Top Twenty Singers') 
plt.ylabel('Number of Songs') 
plt.title('Number of Songs by Singer') 
plt.xticks(rotation=90)
plt.show()
```

```python
liked_count = len(liked_songs)
unliked_count = len(unliked_songs)
```

```python
labels = ['Liked Songs', 'Unliked Songs']
sizes = [liked_count, unliked_count]
colors = ['orange', 'lightgreen']
explode = (0.1, 0)  # explode the first slice (Liked Songs)

# Plotting the pie chart 
plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode,labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, shadow=True)
plt.title('Distribution of Liked and Unliked Songs')
plt.show()
```

```python
# Plot box plot
plt.figure(figsize=(10, 6))
plt.boxplot([liked_songs['duration_ms'], unliked_songs['duration_ms']], labels=['Liked', 'Unliked'], patch_artist=True)
plt.xlabel('Likeability')
plt.ylabel('Duration (ms)')
plt.title('Duration vs Likeability')
plt.show()
```

```python
# Calculate average danceability for each singer
singer_danceability = liked_songs.groupby('singer')['danceability'].mean()
singer_popularity = liked_songs['singer'].value_counts()

# Plot scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(singer_danceability, singer_popularity, color='red', edgecolor='black')
plt.xlabel('Average Danceability')
plt.ylabel('Popularity (Number of Liked Songs)')
plt.title('Danceability vs Popularity of Artists')
plt.show()
```

```python
# Plot histograms for tempo
plt.figure(figsize=(10, 6))
plt.hist(liked_songs['tempo'], bins=30, alpha=0.5, label='Liked Songs', color='orange', edgecolor='black')
plt.hist(unliked_songs['tempo'], bins=30, alpha=0.5, label='Unliked Songs', color='skyblue', edgecolor='black')
plt.xlabel('Tempo (BPM)')
plt.ylabel('Count')
plt.title('Tempo vs Likeability')
plt.legend()
plt.show()
```

```python
# Plot histograms
plt.figure(figsize=(10, 6))
plt.hist(liked_songs['energy'], bins=30, alpha=0.5, label='Liked Songs', color='orange', edgecolor='black')
plt.hist(unliked_songs['energy'], bins=30, alpha=0.5, label='Unliked Songs', color='blue', edgecolor='black')
plt.xlabel('Energy')
plt.ylabel('Count')
plt.title('Energy vs Likeability')
plt.legend()
plt.show()
```
